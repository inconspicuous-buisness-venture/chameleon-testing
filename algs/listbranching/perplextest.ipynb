```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load GPT-2 model and tokenizer
gpt2_model = GPT2LMHeadModel.from_pretrained("gpt2")
gpt2_tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# Function to calculate perplexity
def calculate_perplexity(text):
    tokens = gpt2_tokenizer(text, return_tensors="pt", truncation=True, max_length=512)
    with torch.no_grad():
        outputs = gpt2_model(**tokens, labels=tokens["input_ids"])
    loss = outputs.loss
    perplexity = torch.exp(loss)
    return perplexity.item()

# Example usage
text1 = "Once upon a time, there existed an organization"
text2 = "Once upon a time a long long long long"
text3 = "Once upon a time in America’s heartlands"

print("Perplexity for text1:", calculate_perplexity(text1))
print("Perplexity for text2:", calculate_perplexity(text2))
print("Perplexity for text3:", calculate_perplexity(text3))
```

```python
def generate_and_score(text, max_length=50):
    # Tokenize and generate continuation
    tokens = gpt2_tokenizer(text, return_tensors="pt", truncation=True, max_length=512)
    output = gpt2_model.generate(**tokens, max_length=len(tokens["input_ids"][0]) + max_length, num_return_sequences=1)
    generated_text = gpt2_tokenizer.decode(output[0], skip_special_tokens=True)
    
    # Calculate perplexity for the generated text
    perplexity = calculate_perplexity(generated_text)
    return generated_text, perplexity

# Example usage
text_fragment = "Once upon a time, there was an island where"
generated_text, perplexity = generate_and_score(text_fragment)
print("Generated Text:", generated_text)
print("Perplexity:", perplexity)
```
