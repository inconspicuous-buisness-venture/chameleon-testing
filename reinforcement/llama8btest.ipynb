```python
from transformers import LlamaForCausalLM, AutoTokenizer
import torch
import os

model_path = "E:\\models\\Meta-Llama-3.1-8B"

print(model_path)
tokenizer = None
model = None

try:
    # Load the tokenizer and model with device_map="auto" to manage memory automatically
    tokenizer = AutoTokenizer.from_pretrained(model_path)
    model = LlamaForCausalLM.from_pretrained(model_path, torch_dtype=torch.float16, device_map="auto")
    print("Model and tokenizer loaded successfully.")
except Exception as e:
    print(f"Error loading model or tokenizer: {e}")

if model is not None and tokenizer is not None:
    # Skip the model.to(device) part, as device_map="auto" already handles device placement
    input_text = "Once upon a time"
    inputs = tokenizer(input_text, return_tensors="pt").to(model.device)  # Ensure inputs are on the same device as the model
    outputs = model.generate(**inputs, max_new_tokens=50)

    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    print(generated_text)
else:
    print("Model or tokenizer not available. Aborting text generation.")
```
