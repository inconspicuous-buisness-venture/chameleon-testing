```python
import matplotlib.pyplot as plt
from collections import Counter
import seaborn as sns

# Function to compute word frequencies
def get_word_frequencies(text):
    words = text.lower().split()
    word_counts = Counter(words)
    return list(word_counts.values())

# Function to overlay KDE plots
def overlay_kde_plots(human_texts, ai_texts, title):
    plt.figure(figsize=(10, 6))
    
    # Human plots in red
    for i, text in enumerate(human_texts):
        freq_values = get_word_frequencies(text)
        sns.kdeplot(freq_values, fill=True, color=plt.cm.Reds(0.25 + i / len(ai_texts)), alpha=0, label=f'Human {i + 1}')
    
    # AI plots in blue
    for i, text in enumerate(ai_texts):
        freq_values = get_word_frequencies(text)
        sns.kdeplot(freq_values, fill=True, color=plt.cm.Blues(0.25 + i / len(ai_texts)), alpha=0, label=f'AI {i + 1}')
    
    # Add title, labels, and legend
    plt.title(title)
    plt.xlabel('Frequency')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

overlay_kde_plots(human_text, ai_text, 'KDE of Word Frequencies for Humans and AI')
```
