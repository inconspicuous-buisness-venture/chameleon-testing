```python
import matplotlib.pyplot as plt
from collections import Counter
import seaborn as sns

# Function to compute word frequencies
def get_word_frequencies(text):
    words = text.lower().split()
    word_counts = Counter(words)
    return list(word_counts.values())

# Function to overlay KDE plots for AI texts
def overlay_kde_plots(ai_texts, title):
    plt.figure(figsize=(10, 6))
    
    # Use distinct shades of blue for AI texts
    colors = sns.color_palette("Blues", len(ai_texts))
    
    for i, (temperature, text) in enumerate(ai_texts):
        freq_values = get_word_frequencies(text)
        sns.kdeplot(freq_values, fill=True, color=colors[i], alpha=0, label=f'Temp. {temperature}', clip=(1, None))
    
    # Add title, labels, and legend
    plt.title(title)
    plt.xlabel('Word Frequency')
    plt.ylabel('Density')
    plt.xticks([])  # Remove numbers on the X-axis
    plt.legend()
    plt.show()

# Call the function
overlay_kde_plots(ai_texts, 'KDE of Word Frequencies for AI Texts')
```
