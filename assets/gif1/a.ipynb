```python
from PIL import Image
import os

# Path to the folder containing the images
folder_path = 'branch_frames'

# Get all image file names in the folder
image_files = [f for f in os.listdir(folder_path) if f.endswith(('png', 'jpg', 'jpeg', 'gif'))]
image_files.sort()  # Sort the images by name

# Load images into a list
images = [Image.open(os.path.join(folder_path, file)) for file in image_files]

# Save as a GIF
output_path = 'branch_animation.gif'
images[0].save(output_path, save_all=True, append_images=images[1:], loop=0, duration=100)

print(f"GIF saved at {output_path}")
```
