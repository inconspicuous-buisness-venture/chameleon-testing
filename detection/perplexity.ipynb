```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import torch

# Load pre-trained GPT-2 model and tokenizer
model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Put the model in evaluation mode
model.eval()

def calculate_perplexity(text):
    inputs = tokenizer.encode(text, return_tensors="pt")
    with torch.no_grad():
        outputs = model(inputs, labels=inputs)
        loss = outputs.loss
        perplexity = torch.exp(loss)  # Compute perplexity
    return perplexity.item()

def detect_ai_generated_text(text, perplexity_threshold=50):
    perplexity = calculate_perplexity(text)
    print(f"Text Perplexity: {perplexity}")
    
    # A lower perplexity means the text is more likely AI-generated
    if perplexity < perplexity_threshold:
        return "AI-Generated Text"
    else:
        return "Likely Human-Written Text"
```

```python
# Example text
text = """
The automobile—an entity at once mechanical and existential—inhabits a strange liminal space between utility and symbol, a rolling paradox that encapsulates both freedom and constraint. To unravel the car’s tangled evolution is to embark on an odyssey through not only technological advancement but also the intricacies of human desire, societal structure, and the metaphysics of motion itself.
"""
result = detect_ai_generated_text(text)
print(f"Detection Result: {result}")
```
