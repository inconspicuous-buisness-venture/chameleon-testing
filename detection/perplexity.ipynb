```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import torch

model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

model.eval()

def calculate_perplexity(text):
    inputs = tokenizer.encode(text, return_tensors="pt")
    with torch.no_grad():
        outputs = model(inputs, labels=inputs)
        loss = outputs.loss
        perplexity = torch.exp(loss)
    return perplexity.item()

def detect_ai_generated_text(text, perplexity_threshold=50):
    perplexity = calculate_perplexity(text)
    print(f"Text Perplexity: {perplexity}")
    
    if perplexity < perplexity_threshold:
        return "AI-Generated Text"
    else:
        return "Likely Human-Written Text"
```

```python
# Example text
text = """
This project seeks to determine the contribution of sediment-derived Dissolved Organic Carbon (DOC) to surface water Colored Dissolved Organic Matter (CDOM) in coastal regions. Specifically, the study will focus on the Texas-Louisiana shelf system to assess the extent to which sediment-derived DOC can be identified and quantified using remote sensing technologies. This information will enhance the understanding of coastal carbon budgets, which are critical for mitigating climate change impacts.
Dissolved organic carbon (DOC) fluxes from coastal sediments are globally significant, contributing to the carbon budget of ocean systems. However, there is limited understanding of how sediment-derived DOC influences surface waters, especially in coastal regions with complex hydrodynamics, like the Texas-Louisiana shelf. Studies such as Burdige et al. (1992) and Signorini et al. (2019) have demonstrated the importance of sediment DOC fluxes and their connection to surface carbon pools. Yet, few studies have systematically identified sediment-derived DOC using remote sensing due to physical disconnects between sediments and surface water visible to satellites, and the confounding influence of fluvial inputs.
"""
result = detect_ai_generated_text(text)
print(f"Detection Result: {result}")
```
