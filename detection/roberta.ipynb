```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("openai-community/roberta-base-openai-detector")
model = AutoModelForSequenceClassification.from_pretrained("openai-community/roberta-base-openai-detector")

def detect_gpt_generated_text(text):
    # Tokenize the input text
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)

    # Perform inference
    with torch.no_grad():
        outputs = model(**inputs)

    # Get the logits (output scores before activation)
    logits = outputs.logits
    # Apply softmax to get probabilities
    probabilities = torch.softmax(logits, dim=-1)
    
    # Index 0: Not GPT-generated, Index 1: GPT-generated
    prob_gpt_generated = 1 - probabilities[0][1].item()
    
    return prob_gpt_generated
```

```python
# Example usage
text = """
Cars, intricate marvels of engineering, have long fascinated and fundamentally reshaped the dynamics of human existence, spanning from the cobbled streets of the 19th century to the cutting-edge superhighways of today. These once-clunky, steam-driven oddities, scarcely imagined as anything more than a luxury for the affluent, metamorphosed into essential companions for millions, courtesy of pioneering mass-production techniques like Ford's assembly line—a revolution that turned personal mobility into an attainable dream. Each decade brought forth seismic shifts in automotive design and technology: from the growling muscle cars that echoed freedom and rebellion to the sleek, computer-laden electric vehicles that glide silently today. The internal combustion engine, a roaring symbol of industrial might, now shares its stage with whisper-quiet electric motors, as a looming global urgency around climate change spurs the rapid evolution toward a more sustainable future. But a car is not merely a utilitarian tool; it's a canvas for identity, a status symbol, an enabler of exploration, and often a space for introspective solitude. In this era of unprecedented connectivity, the modern vehicle is as much a data-driven device as it is a transport tool, laden with sensors, cameras, and artificial intelligence systems—transforming how we interact with our surroundings. As autonomous driving technology inches toward mainstream adoption, with vehicles poised to navigate cities on their own, the concept of driving itself is being redefined, blurring the lines between human agency and machine autonomy. This trajectory, while brimming with promise, poses myriad philosophical, ethical, and logistical questions about safety, privacy, and the very fabric of our transportation infrastructure. The future of cars, it seems, will not just be one of technological sophistication but of profound societal transformation, as the journey from point A to point B becomes ever more intertwined with the digital, ecological, and autonomous revolutions of the 21st century.
"""
probability = detect_gpt_generated_text(text)

print(f"Probability that the text was generated by GPT: {probability * 100:.2f}%")
```
